```python
# national park fee 
from matplotlib import pyplot as plt
plt.style.use('ggplot')


# read csv files 
df_np_fee = pd.read_csv('/Users/hayley/workspace/scrapyProject/NationalParkEntranceFee/entranceFee.csv',encoding="ISO-8859-1")

#data cleansing time column

def convert_date(time):
    result = time
    
    #1.Jan.19 -> 20190119
    if(time=='Current'):
        result = '20181231' 
        
    else:
        dict_month = {"Jan": "01", "May": "05"}
        tmp = ('20',time.split(".")[2],dict_month[time.split(".")[1]],'0',time.split(".")[0])
        result = "".join(tmp)
   
    return result


#df_np_fee["time"].apply(lambda x: x.replace(',','')).astype('int')
df_np_fee["time"] = df_np_fee["time"].apply(convert_date)

df_np_fee = df_np_fee[df_np_fee["time"] < '20190720'] 
df_np_fee["tmp"] = df_np_fee.groupby(['park_name'])['time'].transform(max)
df_np_fee = df_np_fee[df_np_fee["time"] == df_np_fee["tmp"]] 

df_np_fee = df_np_fee.drop_duplicates()

df_np_fee['per_person'] = df_np_fee['per_person'].apply(lambda x: (x.replace('$','')) or (x.strip().replace('',0))).astype('int')
#df_np_fee['annual_pass'] = df_np_fee['annual_pass'].apply(lambda x: (x.replace('$','')) or (x.replace(' ',0)))
df_np_fee['annual_pass'] = df_np_fee['annual_pass'].apply(lambda x: (x.replace('$','')) or (x.strip().replace('',0)))

# park name editing
df_np_fee['park_name'] = df_np_fee['park_name'].apply(lambda x: x.replace('National Park','NP'))

#Great Smoky Mountains fee is free. Insert Great Smoky Mountains's data!
df_greatsmoky = pd.DataFrame({'annual_pass':[0],
                              'park_name': ['Great Smoky Mountains NP'],
                              'per_motorcycle': [0], 
                              'per_person' : [0], 
                              'per_vehicle': [0],
                              'time' : ['20181231'],
                              'tmp':['20181231']})
                             #[[647,0,'Great Smoky Mountains NP',0,0,0,'20181231','20181231']])

df_np_fee = pd.concat([df_np_fee,df_greatsmoky], axis=0)
#print(df_np_fee)

#top10_df_np_fee = df_np_fee[df_np_fee["park_name"].isin(np_name)]
top10_df_np_fee = df_np_fee[df_np_fee["park_name"].isin(np_name)]

top10_df_np_fee.set_index("park_name",inplace=True)

top10_df_np_fee['annual_pass'] = top10_df_np_fee['annual_pass'].astype('int')

#top10_df_np_fee.drop('tmp', axis=1, inplace=True)

top10_df_np_fee = top10_df_np_fee.sort_values('per_person',ascending=False)

# draw graph 
plot_top10_np_fee = top10_df_np_fee.plot(kind='barh',figsize=(8,6), color=plt.cm.Paired(np.arange(len(top10_df_np_fee))));

#plt.plot(top10_df_np_fee,kind='bar')
#plt.plot(top10_df_np_fee,kind='bar',stacked=True,figsize=(10,7), color=plt.cm.Paired(np.arange(len(top10_df_np_fee))))
    
    
#op10_df_np_fee= top10_df_np_fee['visits_number'].plot(kind='bar',stacked=True,figsize=(10,7), color=plt.cm.Paired(np.arange(len(df_top10_visitors))));

plot_top10_np_fee.set_title("National Parks's Entrance Fee", fontsize=12)
 
## x axis label 
plot_top10_np_fee.set_xlabel("Entrance Fee")
 
# y axis label
plot_top10_np_fee.set_ylabel("Natioanl Park Name")
#plot_top10_visitors.set_yticks(np.arange(0, 12000000, step=1000000))

#plot.show()

'''
# .astype('int')

#df_np_fee.sort_values('per_person',ascending=False)
.

#np_name = ["Acadia NP", "Zion NP", "Glacier NP", "Grand Canyon NP", "Grand Teton NP", "Great Smoky Mountains NP", "Olympic NP", "Rocky Mountain NP", "Yellowstone NP", "Yosemite NP"]

#print(df_np_fee[df_np_fee["park_name"].isin(np_name)])

# draw graph 10 most national parks's fee 

#top10_df_np_fee = df_np_fee[df_np_fee["park_name"].item in np_name]

top10_df_np_fee = df_np_fee[df_np_fee["park_name"].isin(np_name)]


print(top10_df_np_fee.sort_values('park_name',ascending=True))


#set index 
df_np_fee.set_index("park_name",inplace=True)


# draw graph 
plot_top10_visitors= df_top10_visitors['visits_number'].plot(kind='bar',stacked=True,figsize=(10,7), color=plt.cm.Paired(np.arange(len(df_top10_visitors))));

plot_top10_visitors.set_title("10 Most Visited National Parks in 2018", fontsize=20)
 
## x axis label 
plot_top10_visitors.set_xlabel("Natioanl Park Name")
 
# y axis label
plot_top10_visitors.set_ylabel("Visitors")
plot_top10_visitors.set_yticks(np.arange(0, 12000000, step=1000000))
'''


top10_df_np_fee
```
